```python
def compute_hr(sig_len,qrs_ins,fs):
    heart_rate=np.full(sig_len,np.nan,dtype="float32")
    
    if len(qrs_inds)<2:
        return heart_rate
    
    for i in range(0,len(qrs_inds)-2):
        a=qrs_inds[i]
        b=qrs_inds[i+1]
        c=qrs_inds[i+2]
        rr=(b-a)*(1.0/fs)*1000
        hr=60000.0/rr
        heart_rate[b+1:c+1]=hr
        
        heart_rate[qrs_inds[-1]:]=heart_rate[qrs_inds[-1]]
        
        return heart_rate
```

```python
def compute_hr_custom(r_peak_1,r_peak_2,fs):
    a=r_peak_1
    b=r_peak_2
    rr=(b-a)*(1.0/fs)
    hr= 60/rr
    
    return hr
```

```python
#获取所需数据
for number in range(1,10):
    record_ecg= wfdb.rdrecord('../tset01/physionet.org/files/picsdb/1.0.0/infant{}_ecg'.format(number))
    record_resp=wfdb.rdrecord('../tset01/physionet.org/files/picsdb/1.0.0/infant{}_resp'.format(number))
    r_peaks= wfdb.rdann('../tset01/physionet.org/files/picsdb/1.0.0/infant{}_ecg'.format(number),'qrsc')
    resp_peaks=wfdb.rdann('../tset01/physionet.org/files/picsdb/1.0.0/infant{}_resp'.format(number),'resp')

    record_ecg=record_ecg
    record_ecg=record_ecg
    r_peaks=r_peaks
    resp_peaks=resp_peaks
    
    #calcute heart rate and breathing rate
    heart_rate=processing.compute_hr(len(record_ecg.p_signal),r_peaks.sample,record_ecg.fs)
    breathing_rate=processing.compute_hr(len(record_resp.p_signal),resp_peaks.sample,record_resp.fs)
    
    custom_hr=compute_hr_custom(r_peaks.sample[0],r_peaks.sample[1],record_ecg.fs)
    wfdb_hr=next(x for x in heart_rate if not np.isnan(x))

    print("heart rate obtain from our own custom function:"+str(round(custom_hr,2)))
    print("heart rate obtain from WFDB's compute_hr function:"+str(round(wfdb_hr,2)))
    
    #get the sampling interval
    ecg_samp_interval=1/record_ecg.fs
    resp_samp_interval=1/record_resp.fs
    
    print("ECG sampling {} frequency:".format(number),record_ecg.fs,"HZ")
    print("ECG sampling {} interval every:".format(number),ecg_samp_interval,"sec")
    print("RESP sampling {} frequency:".format(number),ecg_samp_interval,"sec")
    print("RESP sampling {} interval every:".format(number),ecg_samp_interval,"sec")
    
    # get the time axis
    time_ecg=np.arange(record_ecg.p_signal.shape[0])*ecg_samp_interval
    time_resp=np.arange(record_resp.p_signal.shape[0])*resp_samp_interval
    
    #create heart rate and respiratory rate dataframe
    heart_date={
        'Infant:': number,
        'Heart_rate': heart_rate,
        'Time(sec)': time_ecg,
    }

    resp_data={
        'Infant:': number,
        'Heart_rate': breathing_rate,
        'Time(sec)': time_resp,
    }
    heart_df = pd.DataFrame.from_dict(heart_date)
    resp_df = pd.DataFrame.from_dict(resp_data)  
    
    print("Respiratory DF {}:".format(number))
    # resp_df['Age Group'] = resp_df['Age'].apply(age_groups)

    resp_df
    
    
    #drop the NaN rows
    resp_df=resp_df.dropna()
    resp_df=resp_df.dropna().reset_index(drop=True)

    heart_df=heart_df.dropna()
    heart_df=heart_df.dropna().reset_index(drop=True)
    
    #merge the two datafram together with the respect to the same Time(sec) colum
    merged_df=pd.merge(heart_df,heart_df,how='inner',on=None)
    merged_df
    
    number = number+1
    
```
