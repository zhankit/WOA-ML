```python
# Demo 19 - Use the GQRS detection algorithm and correct the peaks

def peaks_hr(sig, peak_inds, fs, title, figsize=(20, 10), saveto=None):
    #"Plot a signal with its peaks and heart rate"
    # Calculate heart rate
    hrs = processing.hr.compute_hr(sig_len=sig.shape[0], qrs_inds=peak_inds, fs=fs)   
    N = sig.shape[0]
    plt.show()

# Load the WFDB record and the physical samples
data_dir = "../tset01/physionet.org/files/picsdb/1.0.0"
record = wfdb.rdrecord(f'{data_dir}/infant1_ecg', sampfrom=0, sampto=10000, channels=[0])

denosie_record_p_signal=denoise(record.p_signal)
denosie_record_fs=denoise(record.fs)

# Use the GQRS algorithm to detect QRS locations in the first channel
qrs_inds = processing.qrs.gqrs_detect(sig=denosie_record_p_signal, fs=denosie_record_fs)

# Plot results
peaks_hr(sig=denosie_record_p_signal, peak_inds=qrs_inds, fs=denosie_record_fs,
         title="GQRS peak detection on record 100")
    
# Correct the peaks shifting them to local maxima
min_bpm = 20
max_bpm = 230
#min_gap = record.fs * 60 / min_bpm
# Use the maximum possible bpm as the search radius
search_radius = int(denosie_record_fs * 60 / max_bpm)
corrected_peak_inds = processing.peaks.correct_peaks(denosie_record_p_signal, 
                                                     peak_inds=qrs_inds,
                                                     search_radius=search_radius, 
                                                     smooth_window_size=150)

# Display results
print('Corrected GQRS detected peak indices:', sorted(corrected_peak_inds))
peaks_hr(sig=denosie_record_p_signal, peak_inds=sorted(corrected_peak_inds), fs=denosie_record_fs,
         title="Corrected GQRS peak detection on sampledata/100")
```
